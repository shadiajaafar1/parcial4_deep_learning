# EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("template_session.txt", delimiter=";")
data.head()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['time index'], data['acc_x'], label='acc_x', alpha=0.7)
plt.plot(data['time index'], data['acc_y'], label='acc_y', alpha=0.7)
plt.plot(data['time index'], data['acc_z'], label='acc_z', alpha=0.7)
plt.title('Aceleraciones en función del tiempo')
plt.xlabel('Time Index')
plt.ylabel('Acceleration (m/s^2)')
plt.legend()
plt.grid()
plt.show()
```

- acc_x (línea azul): Se mantiene constante alrededor de -4.5m/s, con ligeras fluctuaciones, lo que indica estabilidad en el movimiento a lo largo del eje x.

- acc_y (línea naranja): Tiene valores más bajos, alrededor de 7.5 
 , y muestra pequeñas variaciones, lo que podría implicar un componente de aceleración más estable en este eje.

- acc_z (línea verde): Se encuentra cerca de -4.5 m/s, similar a acc_x, pero parece mostrar más variabilidad en ciertos puntos, lo que sugiere posibles cambios dinámicos en el eje x.

Este gráfico sugiere un sistema con aceleraciones mayormente constantes a lo largo del tiempo, con algunas oscilaciones menores en los ejes acc_x y acc_z.

```python
plt.figure(figsize=(10, 6))
plt.hist(data['acc_x'], bins=15, alpha=0.5, label='acc_x')
plt.hist(data['acc_y'], bins=15, alpha=0.5, label='acc_y')
plt.hist(data['acc_z'], bins=15, alpha=0.5, label='acc_z')
plt.title('Distribución de aceleraciones')
plt.xlabel('Acceleration (m/s^2)')
plt.ylabel('Frequency')
plt.legend()
plt.grid()
plt.show()
```

El eje acc_y muestra menor variabilidad, lo que sugiere un comportamiento más constante en este eje. Por otro lado, acc_x y acc_z tienen distribuciones similares, aunque acc_z parece mostrar mayor dispersión. Además, Estas distribuciones pueden indicar la presencia de fuerzas gravitacionales predominantes o patrones específicos en el sistema medido.

```python
plt.figure(figsize=(10, 6))
plt.plot(data['time index'], data['gyr_x'], label='gyr_x', alpha=0.7)
plt.plot(data['time index'], data['gyr_y'], label='gyr_y', alpha=0.7)
plt.plot(data['time index'], data['gyr_z'], label='gyr_z', alpha=0.7)
plt.title('Velocidades angulares en función del tiempo')
plt.xlabel('Time Index')
plt.ylabel('Gyroscope (rad/s)')
plt.legend()
plt.grid()
plt.show()
```

Todas las componentes (x, y, z) parecen oscilar alrededor de 0, lo cual indica que las velocidades angulares promedio en cada eje no tienen un sesgo significativo hacia una dirección específica. Hay variaciones en los valores de cada eje, con picos positivos y negativos, indicando movimientos rotacionales en diferentes direcciones en cada momento. En algunos puntos (e.g., cerca del índice 1000), se observan picos más pronunciados, lo que podría indicar movimientos rápidos o bruscos en el sistema medido.

```python
plt.figure(figsize=(10, 6))
plt.plot(data['time index'], data['mag_x'], label='mag_x', alpha=0.7)
plt.plot(data['time index'], data['mag_y'], label='mag_y', alpha=0.7)
plt.plot(data['time index'], data['mag_z'], label='mag_z', alpha=0.7)
plt.title('Campo magnético en función del tiempo')
plt.xlabel('Time Index')
plt.ylabel('Magnetic Field (μT)')
plt.legend()
plt.grid()
plt.show()
```

mag_y tiene el valor más alto, cercano a 0.7 µT, y es bastante estable, indicando un predominio del campo magnético en esta dirección. Por su parte, mag_z se mantiene alrededor de 0.2 µT, también con estabilidad pero con ligeras variaciones.Por últimos, mag_x tiene los valores más bajos, cercanos a 0.0 µT, sugiriendo que la contribución del campo magnético en esta dirección es mínima.
